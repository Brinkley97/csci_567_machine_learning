# Linear Regression Plot

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import poisson, norm
```

```python
# plt.style.use('_mpl-gallery')

# make data
n = 5
X = poisson.rvs(1, size=n)
# print(X)
eps = norm.rvs(size=n)
# print(eps)

Y = 1 + (3 * X) + eps
print(len(Y))

# plot
fig, ax = plt.subplots()

ax.plot(X, Y)

# ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
#        ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(0)

# Number of samples
num_samples = 500

# Generate X from a Poisson distribution
X = np.random.poisson(lam=1, size=num_samples)

# Generate eps from a random normal distribution
eps = np.random.normal(loc=0, scale=1, size=num_samples)

# Calculate Y
Y = 1 + 3*X + eps

# Plot the data
plt.scatter(X, Y, alpha=0.5)

# conditional mean
plt.plot(X, 1 + 3*X, color='red', label='abline(1, 3)')

plt.title('Scatter Plot of Y vs X with Red Line')
plt.xlabel('X (Poisson distributed)')
plt.ylabel('Y (1 + 3*X + eps)')
plt.show()
```
